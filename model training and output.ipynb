```python
import numpy as np 
import matplotlib.pyplot as mtp  
import pandas as pd  
from sklearn.model_selection import train_test_split  
from sklearn.tree import DecisionTreeClassifier
# Importing the dataset  
df = pd.read_csv('train.csv')
# df.dropna()
df['Rating'].fillna(df['Rating'].median(),inplace=True)
df['star_5f'].fillna(df['star_5f'].median(),inplace=True)
df['star_1f'].fillna(df['star_1f'].median(),inplace=True)
test_data = pd.read_csv('test.csv')
test_data.drop(test_data.columns[1], axis=1, inplace=True)
test_data
test_data.drop(test_data.columns[2:4], axis=1, inplace=True)
# test_data = test_data.fillna(value = {'Rating' :4, 'star_5f' :156, 'star_1f' :29})
test_data = test_data.fillna(value = {'Rating' :4.011089069629038, 'star_5f' :1557.443199381762, 'star_1f' :270.3977856860419})
# test_data = test_data.fillna(value = {'Rating' :4.1, 'star_5f' :4.0, 'star_1f' :0})
df.drop(df.columns[1], axis=1, inplace=True)
df.drop(df.columns[2:4], axis=1, inplace=True)
df=df.dropna()
df1=df["Offer %"].str.replace("%","")
df1
df['Offer %'] = df1
# create test_X which comes from test_data but includes only the columns you used for prediction.
# The list of columns is stored in a variable called features
 
y = df.price1

# Create X (After completing the exercise, you can return to modify this line!)
features = ['id', 'Rating', 'actprice1', 'norating1',
       'noreviews1', 'star_5f', 'star_4f', 'star_3f', 'star_2f', 'star_1f',
       'fulfilled1']
test_X = test_data[features]
# Select columns corresponding to features, and preview the data
X = df[features]
X.head()

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)

# Feature Scaling  
  
# from sklearn.linear_model import LogisticRegression
# from sklearn.metrics import accuracy_score  
# model=LogisticRegression() 
# model.fit(train_X,train_y) 
# from sklearn.preprocessing import StandardScaler  
# sc = StandardScaler()  
# train_X = sc.fit_transform(train_X)  
# test_X = sc.transform(test_X)  


model= DecisionTreeClassifier(class_weight=None, criterion='entropy', max_depth=100,
 max_features=None, max_leaf_nodes=None,
  min_samples_leaf=1,
 min_samples_split=2, min_weight_fraction_leaf=0.0,
 random_state=0, splitter='best')




model.fit(train_X,train_y)
# make predictions which we will submit. 
test_preds = model.predict(test_X)

# from sklearn.trees import DecisionTreeClassifier
# dtree = DecisionTreeClassifier(max_depth=10 ,random_state=101,max_features=None,min_samples_leaf=15)
# dtree.fit(train_X,train_y)
# test_preds = dtree.predict(test_X)
# from sklearn.ensemble import RandomForestClassifier
# rfm = RandomForestClassifier(n_estimators=5000 ,random_state=0,max_features=None,min_samples_leaf=1,n_jobs=-5)
# rfm.fit(train_X,train_y)
# test_preds = rfm.predict(test_X)

output = pd.DataFrame({'id': test_data.id,'price1': test_preds})


output.to_csv('final3.csv', index=False)

df1 = pd.read_csv('final3.csv')
df1




```
